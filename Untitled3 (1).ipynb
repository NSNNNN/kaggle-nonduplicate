```python
# نصب کتابخانه‌های لازم (مهم!)
!pip install seaborn scikit-learn matplotlib -q
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
print("✅ همه کتابخونه‌ها آماده!")
```

```python
print("=== ۱. فیچرهای دیتاست ===")
print(f"تعداد ردیف‌ها و ستون‌ها: {df.shape}")
print(f"نام فیچرها: {list(df.columns)}")
print("\nآمار توصیفی:")
print(df.describe())
print("\nنوع داده‌ها:")
print(df.dtypes)
```

```python
# فقط فیچرهای عددی
numeric_df = df.select_dtypes(include=[np.number])

plt.figure(figsize=(10,8))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=0.5)
plt.title("ماتریس همبستگی فیچرها")
plt.tight_layout()
plt.show()

print("بالاترین همبستگی:")
print(corr_matrix.abs().unstack().sort_values(ascending=False).drop_duplicates())
```

```python
# PCA فقط برای فیچرهای عددی
numeric_df = df.select_dtypes(include=[np.number]).fillna(0)

scaler = StandardScaler()
data_scaled = scaler.fit_transform(numeric_df)

pca = PCA()
pca.fit(data_scaled)
var_ratio = pca.explained_variance_ratio_
cum_var = np.cumsum(var_ratio)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(range(1, len(var_ratio)+1), var_ratio, 'bo-')
plt.title('واریانس هر کامپوننت')
plt.xlabel('شماره PC')

plt.subplot(1,2,2)
plt.plot(range(1, len(cum_var)+1), cum_var, 'ro-')
plt.axhline(y=0.95, color='g', linestyle='--')
plt.title('واریانس تجمعی')
plt.xlabel('تعداد PC')

plt.tight_layout()
plt.show()

print(f"✅ با {np.argmax(cum_var >= 0.95)+1} فیچر، ۹۵% واریانس پوشش داده می‌شه")
```
